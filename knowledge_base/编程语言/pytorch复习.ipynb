```python
import torch
```

```python
# pytorch直接创建一个一维的张量,值是从0到给定的数字
x = torch.arange(12)
```

```python
# 把张量转换为一维
x.reshape(-1)
```

```python
# 生成全0或全1的张量
full_one = torch.ones(2,1)
full_zero = torch.zeros(3,2)
print(full_one)
print(full_zero)
```

```python
# 通过列表或者嵌套列表构建张量
from_list = torch.tensor([[2,1,3],[4,6,1]])
from_list
```

```python
# 张量支持+,-,*,/,**等运算,运算时会逐元素计算
target = from_list + from_list
print(target)
target = from_list * from_list
print(target)
target = from_list ** from_list
print(target)
```

```python
# 当进行运算的两个张量形状不同时,pytorch会使用广播机制将两个张量对齐为同一形状
# 要使用广播机制,pytorch会从最右边的维度开始检查,每个维度必须满足三个条件之一
# 1.相等
# 2.其中一个为1
# 3.其中一个不存在（即维度较少，缺失的维度自动视为 1）
test1 = torch.tensor([[2,4,1]])
test2 = torch.tensor([[1],[2],[3]])
target = test1 + test2
target
```

```python
# 可以使用cat()函数拼接两个张量
test3 = torch.arange(9).reshape(3,3)
torch.cat((test3,target),dim=0),torch.cat((test3,target),dim=1)
```

```python
# 使用stack对向量进行堆叠,cat()与stack()的不同之处在于,stack会得到更多的维度
torch.stack((test3, target), dim = 0) #新增一个用于区分两个test3和target的维度
```

```python
# 使用cat()时要保证拼接的维度相同,当指定dim=0时就必须保证两个拼接的张量的一行有同样多的元素
torch.cat((test3,test1),dim=0)
```

```python
# 对所有元素求和
target.sum()
```

```python
# 将torch张量和numpy数据类型转换
A = target.numpy()
B = torch.tensor(A)
print(type(A))
print(type(B))
# 将只有一个元素的张量和python的基本类型转换
C = torch.tensor(3.5)
C.item(),float(C)
```

# 自动求导

```python
import torch
```

```python
x = torch.tensor([[2.,3.,4.],[1.,2.,3.]],requires_grad = True) # x.requires_grad = True等价于在创建时后面加requires_grad = True,必须是浮点数或者复数才能使用rad
x.grad
```

```python
y = x ** 2
y
```

```python
y.sum().backward() # backward函数只能对标量使用
x.grad
```

```python
a = torch.tensor([1.,2.,3.])
a.requires_grad = True
z = 2 * torch.dot(a, a) #dot函数只能对一维张量使用,输出两个张量相乘后求和
z.backward() # 只有使用了backward()后梯度才会更新
x.grad
```

```python
a.grad == 4 * a
```

```python
# 在默认情况下,grad会累积梯度我们需要手动清零
y = a.sum().backward()
a.grad == (4 * a + torch.ones_like(a)) 
```

```python
a.grad.zero_() # 手动清零
```

```python
# 使用detach()将计算移动到记录的计算图之外
y = a * a
u = y.detach() #移除y之前的计算
z = u * a
z.sum().backward()
a.grad == u
```

# nn.Module
pytorch的Module子类创建的参数都会自动注册到.parameters()中\
使用标准层会自动包装参数,如果要注册参数需要使用nn.Parameter(tensor)\
通过register_buffer()注册的参数不会更新\
子类一般会重写__init__()函数和forward()函数,nn.Sequential()函数实际相当于流水线的forward()函数\
当使用.eavl()或者.train()切换模型状态时会递归地运用于所有地子模块\
当使用.to(device)或者.float()来改变参数位置或者类型时也会递归地改变

```python
import torch
import torch.nn as nn
net = nn.Sequential(nn.Linear(20, 256), nn.ReLU(), nn.Linear(256, 10)) #线性层第一个维度视作透明,只会在第二个维度计算
X = torch.rand(2, 20)
print(net(X))
```

```python
#只有使用nn.Sequential()这样的可访问下标的创建方式才能这样写,如果是按创建类的方式创建module想这样访问就需要重写__getitem__()
print(net[2].state_dict()) 
print(net[2].bias) #访问最后一层的偏置
print(net[2].bias.data)
```

```python
class MyModule1(nn.Module):
    def __init__(self):
        '''
        init函数用于初始化模块的参数和层
        '''
        super().__init__() #init()函数一定要继承父类地init()函数
        self.hidden = nn.Linear(20, 256) #标准层中地参数会自动注册
        self.activation = nn.ReLU()
        self.out = nn.Linear(256, 10)
        self.para = nn.Parameter(torch.rand(2, 256)) #手动注册一个参数
        self.register_buffer("mask",torch.ones(2, 256)) #注册一个不会变的参数一般掩码就是不需要更新的参数

    
    def forward(self, x):
        '''
        forward参数用于定义计算较为复杂的逻辑
        '''
        x = self.hidden(x)
        x = x * self.para
        x = self.activation(x)
        x = x ** self.mask
        x = self.out(x)
        return x

model = MyModule1()
x = torch.randn(2, 20)
print("=== 输入 ===")
print(x.shape)

print("\n=== 所有可学习参数 ===")
# .parameters()返回的是一个generator对象里面是地址,使用.named_parameters()函数返回generator对象里面包含的是str和tensor
for name, p in model.named_parameters():
    print(f"  {name:20s} | shape: {str(p.shape):18s} | requires_grad: {p.requires_grad}")

print("\n=== state_dict (含buffer) ===")
for k, v in model.state_dict().items():
    print(f"  {k:20s} | shape: {v.shape}")
```

```python
class MySequential(nn.Module):
    def __init__(self, *args):
        "可以通过参数列表来创建"
        super().__init__()
        net = nn.Sequential()
        for i,block in enumerate(args):
            net.add_module(f"block{i}",block) #可以给添加进来的模块命名
        '''
        或者
        for block in args:
            self._modules[block] = block 但这个方法直接访问了_modules不推荐
        '''
        # self.blocks = nn.ModuleList(args) 快速自动注册,但没有命名
            
    def forward(self, x):
        for block in self._modules.values():
            x = block(x)
        return x

net = MySequential(nn.Linear(20, 256), nn.ReLU(), nn.Linear(256, 10))
net(x)
```

```python
# 让不同的层共享参数
x = torch.randn(2,4)
shared = nn.Linear(8, 8)
net1 = nn.Sequential(nn.Linear(4,8),nn.ReLU(),shared,nn.Linear(8,8),shared,nn.ReLU(),nn.Linear(8,1))
net1(x)
print(net1[2].weight == net1[4].weight)
net1[2].weight.data[0,0] = 100
print(net1[2].weight == net1[4].weight)
```

# Dataset and DataLoader

## Dataset类
用于自定义数据集和对数据进行一些数据预处理主要实现三个函数:\
__init__()负责存储元信息（文件路径列表、标签数组、transform 对象）\
__len__()计算数据集的大小即有多少条数据\
__getitem__()函数定义如何加载单条数据以及怎么利用transform进行变换\
## DataLoader
数据加载器,用于封装Dataset,一般有四个比较重要的参数\
batch_size 定义批次大小\
shuffle=True 打乱加载顺序,即不会按照索引顺序依次加载\
num_workers 进程数\
drop_last=True 当使用 BatchNorm 时，尾批 size=1 会导致方差为 0 而 NaN\
collate_fn 定义一个函数用于定制batch的输出,默认的collate_fn会将同一个batch里的数据和标签分别拼接成tuple \

```python
import torch
import time
from torch.utils.data import Dataset,DataLoader
# 对比将数据加载放在init和getitem中的比较
class EagerDataset(Dataset):
    '''
    验证将数据加载放在init中的耗时
    '''
    def __init__(self, n):
        super().__init__()
        start = time.time()
        self.data = [torch.randn(128) for _ in range(n)]
        self.labels = [torch.tensor(i % 10) for i in range(n)]
        print(f"加载耗时{time.time() - start}s")

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return self.data[idx], self.labels[idx]

class LazyDataset(Dataset):
    '''
    getitem()函数是懒加载,需要的内存小于直接在init中加载数据
    '''
    def __init__(self, n):
        super().__init__()
        self.n = n

    def __len__(self):
        return self.n

    def __getitem__(self, idx):
        start = time.time()
        data = torch.randn(128)
        label = torch.tensor(idx % 10)
        print(f"加载耗时{time.time() - start}s")
        return data,label

ds_eager = EagerDataset(5000)
eager_loader = DataLoader(ds_eager, batch_size = 64) #DataLoader的第一个参数都是数据集,其它的参数尽量按关键字传
_ = next(iter(eager_loader)) #触发第一次加载

print("-------------------分隔符----------------------")
ds_lazy = LazyDataset(5000)
lazy_loader =  DataLoader(ds_lazy, batch_size = 64)
_ = next(iter(lazy_loader)) # 必须加上iter才能把DataLoader变成可迭代的
```

```python
# 验证默认collate_fn()函数怎么处理元组、字典和嵌套结构
import torch
from torch.utils.data.dataloader import default_collate

#标准元组
batch_tuple = [(torch.randn(2,3), torch.tensor(i)) for i in range(4)]
print(f"collate 前:{batch_tuple}")
res = default_collate(batch_tuple) #default_collate()函数的底层是stack而非cat函数,使用stack在没有batch维度时会自动创建batch维
print(f"类型: {type(res)}, 长度: {len(res)}")
print(f"data shape: {res[0].shape}, label shape: {res[1].shape}")
print(f"collate 后:{res}")
```

```python
#自定义collate()函数解决变长的形状
class VarlenDataset(Dataset):
    def __init__(self):
        #定义四条变长的数据,长度为3,4,5,2
        self.sentences = [
            torch.tensor([1.,2.,3.]),
            torch.tensor([4.,5.,6.,7.]),
            torch.tensor([8.,9.,10.,11.,12.]),
            torch.tensor([13.,14.])
        ]
        self.labels = [0 ,1 ,2 ,3]

    def __len__(self):
        return len(self.sentences)

    def __getitem__(self, idx):
        return self.sentences[idx], self.labels[idx]

# 自定义collate
def pad_collate(batch):
    '''
    该方法将所有张量的形状对齐为最长的张量形状
    '''
    #batch是一个列表,没一个数据按照(数据,标签)形式排列的,取用不太方便,因为一般都是取用数据进行一些转换操作,会频繁地使用索引来访问
    #当我们将batch中的数据和标签分别堆叠起来能够快速的对数据进行整体的操作
    #zip(*batch)是将batch中的数据取出来再按位置重新分组为[(s1, s2, s3), (l1, l2, l3)]
    sentences, labels = zip(*batch) # 转置: list of tuples → tuple of lists
    #计算当前batch长度最大的张量
    max_len = max(s.size(0) for s in sentences)
    
    #创建跟一个batch大小相等的全0pad
    padded = torch.zeros(len(sentences), max_len, dtype = torch.float)
    lengths = []

    #遍历sentences并填充张量
    for i,s in enumerate(sentences):
        padded[i, : s.size(0)] = s #一维的张量size(0)就是指序列长度,二维张量才是有多少行
        lengths.append(s.size(0))

    return padded, labels, lengths

loader = DataLoader(VarlenDataset(), batch_size = 4, collate_fn = pad_collate)
padded, labels, lengths = next(iter(loader))
print(f"Padded shape: {padded.shape}")  # 预期 [4, 5]
print(f"Padded tensor:\n{padded}")      # 短序列右侧应为0
print(f"Original lengths: {lengths}")   # 预期 [3, 5, 2, 4]
print(f"Labels: {labels}")              # 预期 [0, 1, 0, 1]
```
